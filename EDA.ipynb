## Emotions Dataset Analysis:

## Analysis of train data:

```python
"""
    These are the libraries we will be importing for the Exploratory Data Analysis.
    Pandas to create dataframes and save the paths in a csv format.
    Numpy for array related operations and to assist in plots.
    matplotlib for overall mathematical plots, graphs and analysis.
    seaborn for a stylist approach and analysis of graphs and plots.
    cv2 is opencv for computer vision based applications.
    os for windows system operations.
    We will also include the warnings module to avoid any warnings related to future versions
    or any kind of depreciations.
    
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import cv2 
import os
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline 
sns.set(color_codes=True)
```

```python
"""
    Looking at the contents in the train folder and trying to figure out the total number of classes, 
    that we have for each of the categories for the emotions in train folder.
    
"""

train_directory = "train/"
train_dir = os.listdir(train_directory)
classes = []

for folder in train_dir:
    classes.append(folder)
    print(folder)
```

```python
"""
    We Have a total of 6 emotions namely Angry, Fear, Happy, Neutral, Sad and Surprise.
    In this code cell we run through all the emotions directories and get the counts
    for each of the emotions. 

"""

train_counts = []

for folder in train_dir:
    class_path = train_directory + folder + "/"
    list_train = []
    count = 0
    for file in os.listdir(class_path):
        count +=1
    
    train_counts.append(count)
    
train_counts
```

### Bar Graph:

```python
plt.bar(classes, train_counts, width=0.5)
plt.title("Bar Graph of Train Data")
plt.xlabel("Classes")
plt.ylabel("Counts")
```

### We can notice from the bar graph that this is a fairly balanced dataset

### Scatter Plot:

```python
plt.scatter(classes, train_counts)
plt.plot(classes, train_counts, '-o')
plt.show()
```

### Displot:

```python
x = np.array(train_counts)
sns.distplot(x, kde = True, bins = 5)

sns.distplot(x)
```

### Reading and Analyzing the images:

```python
"""
    In this code block we are reading an image from the train/angry/
    directory where we have one angry image. We will using this particular
    image to analyze the dimensions and the type of image being plotted.

"""

for folder in train_dir:
    class_path = train_directory + folder + "/"
    for file in os.listdir(class_path):
        if file.endswith(".jpg"):
            final_path = class_path + file
            print(final_path)
            img = cv2.imread(final_path, cv2.IMREAD_UNCHANGED)
        break
    break
```

```python
plt.imshow(img)
```

```python
height, width = img.shape

print("""
The Height of the image = {0} 
The Width of the image = {1} 
The Number of channels = {2}""".format(height, width, 1))
```

```python
"""
    Extracting and using the first image file of each directory.

"""

image_paths = []

for folder in train_dir:
    class_path = train_directory + folder + "/"
    for file in os.listdir(class_path):
        if file.endswith(".jpg"):
            final_path = class_path + file
            image_paths.append(final_path)
        break
```

```python
image_paths
```

```python
"""
    We Will use the mpimg module to plot multpile images in a single code 
    block cell. Here we navigate through the first image of every 
    single directory and then take a look at the type of emotions
    we are dealing with.

"""


import matplotlib.image as mpimg

for file in image_paths:
    image = mpimg.imread(file)
    plt.figure()
    pos1 = file.find('/')
    pos2 = file.find('/0')
    title = file[pos1+1:pos2]
    plt.title(title, fontsize = 20)
    plt.imshow(image)
```
